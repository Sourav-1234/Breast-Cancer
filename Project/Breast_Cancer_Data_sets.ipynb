### Data Set Information:

This is one of three domains provided by the Oncology Institute that has repeatedly appeared in the machine learning literature. (See also lymphography and primary-tumor.) 

This data set includes 201 instances of one class and 85 instances of another class. The instances are described by 9 attributes, some of which are linear and some are nominal. 


### Attribute Information:

1. Class: no-recurrence-events, recurrence-events 
2. age: 10-19, 20-29, 30-39, 40-49, 50-59, 60-69, 70-79, 80-89, 90-99. 
3. menopause: lt40, ge40, premeno. 
4. tumor_size: 0-4, 5-9, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40-44, 45-49, 50-54, 55-59. 
5. inv_nodes: 0-2, 3-5, 6-8, 9-11, 12-14, 15-17, 18-20, 21-23, 24-26, 27-29, 30-32, 33-35, 36-39. 
6. node_caps: yes, no. 
7. deg_malig: 1, 2, 3. 
8. breast: left, right. 
9. breast-quad: left-up, left-low, right-up,	right-low, central. 
10. irradiat:	yes, no.

## Converting the irradiat category variable  yes and no to numeric and dropping it

```python
df3["irradiat"]
df3['irradiat_cat'] = df3['irradiat'].map( {'yes':1, 'no':0} )
df3.drop('irradiat', axis=1,inplace = True)
```

## Converting the node_caps category variable yes and no to numeric and dropping it

```python
df3["node_caps"]
df3['node_cat'] = df3['node_caps'].map( {'yes':1, 'no':0} )
df3.drop('node_caps', axis=1,inplace = True)
```

## Converting  the Breast Column category variable  right and left to numeric and dropping it

```python
df3["breast"]
df3['breast_cat'] = df3['breast'].map( {'right':1, 'left':0} )
df3.drop('breast', axis=1,inplace = True)
```

## Decision Tree

```python
from sklearn.model_selection import train_test_split
# Train-Test split: 20% test data

label = df2_dummies.irradiat_cat
data_train, data_test, label_train, label_test = train_test_split(df2_dummies, label, test_size = 0.2, random_state = 50)
#data_test.fillna(data_test.mean())
#data_train.fillna(data_train.mean())
#label_test.fillna(label_test.mean())
#label_train.fillna(label_train.mean())
```

```python
from sklearn import tree
# Decision tree with depth = 2
dt2 = tree.DecisionTreeClassifier()
dt2.fit(data_train, label_train)
dt2_score_train = dt2.score(data_train, label_train)
print("Training score: ",dt2_score_train)
dt2_score_test = dt2.score(data_test, label_test)
print("Testing score: ",dt2_score_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(solver = 'lbfgs', multi_class = 'ovr')

lr_model.fit(data_train, label_train)
```

## Predictor and Test

```python
y_pred = lr_model.predict(data_test)
```

```python
y_pred
```

```python
label_test
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(label_test, y_pred)
```

```python
cm
```

```python
corrects = np.trace(cm)

total = np.sum(cm)
```

## Calculating the Accuracy

```python
print("Correctly identified: {} and total: {}".format(corrects, total))
```

```python
accuracy = corrects/total
```

```python
print("Accuarcy: {}%".format(accuracy*100))
```

## SVM Model

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(data_train, label_train)
```

```python
y_pred = svclassifier.predict(data_test)
```

## Confusion Matrix

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(label_test,y_pred))
print(classification_report(label_test,y_pred))
```
